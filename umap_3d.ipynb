```python
vectors = []
for word in words:
    vectors.append(nlp(word).vector)
vectors = np.array(vectors)
```

```python
n_words = len(words)
words = np.array(words[:n_words])
vectors = vectors[:n_words, :]
```

```python
reducer = umap.UMAP(n_components=3)
scaled_data = StandardScaler().fit_transform(vectors)
```

```python
embedding = reducer.fit_transform(scaled_data)
embedding.shape
```

```python
fig = px.scatter_3d(x = embedding[:, 0], y = embedding[:, 1], z = embedding[:, 2], text = words, hover_name=words,
                   size = [1000 for i in range(n_words)])

fig.show()
```

```python
import plotly.graph_objects as go
```

```python
data = go.Scatter3d(x = embedding[:, 0], y = embedding[:, 1], z = embedding[:, 2], text = words,
                   mode='markers',
                   marker=dict(
        size=20,
        opacity=0.8
    ))
layout = go.Layout(width=1280, height=720)
```
