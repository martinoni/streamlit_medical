```python
texts
```

```python
nlp = spacy.load("en_core_sci_lg")
```

```python
words = []
vecs = []
for text in texts:
    doc = nlp(text)
    for token in doc:
        if token.has_vector:
            words.append(token)
            vecs.append(token.vector)
```

```python
data_vecs = pd.DataFrame(vecs)
```

```python
data_vecs.index = words
```

```python
data_vecs_unique = data_vecs.drop_duplicates()
```

```python
data_vecs_unique
```

```python
words = data_vecs_unique.index
```

```python
import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
import umap
```

```python
reducer = umap.UMAP()
```

```python
scaled_data = StandardScaler().fit_transform(data_vecs_unique)
```

```python
embedding = reducer.fit_transform(scaled_data)
embedding.shape
```

```python
embedding.shape
word2Ind = {str(words[i]):i for i in range(len(words))}
```

```python
plt.rcParams["figure.figsize"] = (40, 40)
fig, ax = plt.subplots()

ax.scatter(
    embedding[:, 0],
    embedding[:, 1])

for word in words:
    word = str(word)
    x, y = embedding[word2Ind[word]]
    plt.text(x+.03, y+.03, word, fontsize=9)
    
plt.gca().set_aspect('equal', 'datalim')
plt.title('UMAP projection of the Penguin dataset', fontsize=24)
```
